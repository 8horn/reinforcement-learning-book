Contrary to Dynamic Programming, with *Monte Carlo* methods we not assume complete knowledge of the environment and only require *experience* - sample sequences of states, actions and rewards from actual or simulated interaction with an environment. Even though the model is required, it only needs to generate sample transitions, not the complete probability distributions of all possible transitions that is required for dynamic programming. 

More generic, **Monete Carlo** methods are ways of solving the reinforcement learning problem based on averaging sample returns. For this purpose, we assume experience is divided into episodes and that all episodes eventually terminate no matter what actions are selected. Only on completion of an episode are value estimates and policies changed. Monte Cardlo methods are thus incremental in an *episode-by-episode* sense, but not in a *step-by-step* (online) sense. 

Monte Carlo methods sample and average *returns* for each state-action pair, much like the bandit methods sample and average *reward* for each action. The difference is that now we are dealing with associative search problem, which is more difficult as there are multiple states and the return after taking an action in one state depends on the actions taken in later states in the same episode. 

To implement and test Monte Carlo methods, we wil use the Blackjack environment

```python
import numpy as np

class Player():
   def __init__(self, goal, available_cards):
      self.goal = goal
      self.available_cards = available_cards
      self.natural = False
      self.sticks = False
      self.bust = False
      
   def __repr__(self):
      return f'Cards: {self.cards}\nNatural: {self.natural}\nSticks: {self.sticks}\nBust: {self.bust}'

   def __call__(self):
      return self.cards.sum()
   
   def reset(self):
      self.natural = False
      self.sticks = False
      self.bust = False
      self.cards = np.random.choice(self.available_cards, 2)
      self.evaluate_aces()
      self.is_natural()

   def evaluate_aces(self):
      ace_indices = np.argwhere(self.cards == 11).reshape(-1)
      if ace_indices.size == 0: 
         # No aces, no actions
         pass
      else:
         for ace_index in ace_indices:
            # If sum without single ace over 11, then ace counts as 1
            no_ace_sum = self.cards[np.arange(len(self.cards)) != ace_index].sum()
            if no_ace_sum >= 11:
               self.cards[ace_index] = 1
   
   def is_natural(self):
      if self.cards.sum() == self.goal: self.natural = True
      
   def is_bust(self):
      if self.cards.sum() > self.goal: self.bust = True
      
   def hit(self):
      new_card = np.random.choice(self.available_cards)
      self.cards = np.append(self.cards, new_card)
      self.evaluate_aces()
      self.is_bust()
      
   def stick(self):
      self.sticks = True
      
   def act(self, action):
      if action == 0: 
         self.hit()
      else: 
         self.stick()
    
      
class Blackjack():
   def __init__(self):
      self.goal = 21
      self.available_actions = [0, 1] # 0 - hit, 1 - stick
      self.available_cards = [2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10, 11] # face cards count as 10 
      self.player = Player(self.goal, self.available_cards)
      self.dealer = Player(self.goal, self.available_cards)
      
   def reset(self):
      self.player.reset()
      self.dealer.reset()
      self.dealer_visible = self.dealer.cards[0]
      
      state = (self.player(), self.dealer_visible)
      reward = 0
      done = False
      
      if self.player.natural or self.dealer.natural: 
         done = True
    
      if self.player.natural and not self.dealer.natural:
         reward = 1
      elif self.dealer.natural and not self.player.natural:
         reward = -1
      
      return state, reward, done
   
   def step(self, player_policy=None, dealer_policy=None):
      action = self.step_player(player_policy)
      next_state = (self.player(), self.dealer_visible)
      reward = 0
      done = False
      
      if self.player.bust:
         reward = -1
         done = True
      if self.player.sticks:
         self.step_dealer(dealer_policy)
         if self.dealer.bust: 
            reward = 1
         else: 
            if self.dealer() > self.player(): reward = -1
            elif self.dealer() < self.player(): reward = 1
            else: reward = 0
         done = True
         
      return next_state, action, reward, done
      
   def step_player(self, player_policy):
      if not player_policy: 
         action = np.random.choice(self.available_actions) 
      elif isinstance(player_policy, (int, float)):
         action = int(player_policy)
      else:
         action = player_policy(self.player())
         
      self.player.act(action)
      return action
      
   def step_dealer(self, dealer_policy):
      while not (self.dealer.bust or self.dealer.sticks):
         if not dealer_policy: 
            action = np.random.choice(self.available_actions) 
         else: 
            action = dealer_policy(self.dealer())
            
         self.dealer.act(action)
```

```python
dealer_policy = lambda x: 1 if x >= 17 else 0
player_policy = lambda x: 1 if x == 20 or x == 21 else 0
```

First, let's consider Monte Carlo methods for learning state-value function for a given policy. Suppose we wish to estimate $v_\pi(s)$, the value of a state $s$ under policy $\pi$, given a set of episodes obtained by following $\pi$ and passing through $s$. **First-visit MC method** estimates $v_\pi(s)$ as the averae of the reutrns following first visits to s, whereas the **every-visit MC method** averages the returns following all visits to $s$. By visit, we consider cases where state $s$ has been visitied in a single episode.

<img src="./resources/first-visit-MC.JPG" />

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from collections import defaultdict

def first_visit_mc(env, player_policy, dealer_policy, gamma=1., n_episodes=1000):
   state_size = (env.goal+1, max(env.available_cards)+1)
   V = np.zeros(state_size)
   returns = defaultdict(list)
   
   for i_episode in range(n_episodes):
      # Generate an episode starting from empty state and store intermediate results in list
      state, reward, done = env.reset()
      action = None
      episode_steps = []
      if done: episode_steps.append((state, action, reward))
      while not done:
         next_state, action, reward, done = env.step(player_policy, dealer_policy)
         episode_steps.append((state, action, reward))
         state = next_state

      # Iterate through the episode steps list in reversed order and calculate expcted returns
      G = 0
      for episode_step in reversed(episode_steps):
         state, action, reward = episode_step
         G = gamma * G + reward
         returns[state].append(G)
         V[state] = np.average(returns[state])
      #print(G, returns[state], np.average(returns[state]), V[state])
         
   return V, returns

def plot_expected_reward(V, title):
   x = np.linspace(1, V.shape[1], V.shape[1])
   y = np.linspace(12, V.shape[0], V[12:, :].shape[0])
   X, Y = np.meshgrid(x, y)

   fig = plt.figure(figsize=(15, 10))
   ax = plt.axes(projection='3d')
   ax.plot_surface(X, Y, V[12:, :], rstride=1, cstride=1, cmap='BuGn')

   ax.set_xlabel('Dealer Showing')
   ax.set_ylabel('Player Sum')
   ax.set_zlabel('Expected Reward')
   ax.set_title(title)

   plt.show()
```

```python
bj = Blackjack()

dealer_policy = lambda x: 1 if x >= 17 else 0
player_policy = lambda x: 1 if x == 20 or x == 21 else 0
```

```python
V, _ = first_visit_mc(bj, player_policy, dealer_policy, n_episodes=10000)
```

```python
plot_expected_reward(V, 'After 10,000 episodes')
```

```python
V, _ = first_visit_mc(bj, player_policy, dealer_policy, n_episodes=500000)
```

```python
plot_expected_reward(V, 'After 500,000 episodes')
```

### Monte Carlo Estimation of Action Values

With a model, state values alone are sufficient to determina a policy, but when model is not available, then we might want to estimate values of each *action*, so that it is useful for suggesting improved policies. 

The polict evaluation problem for action values is to estimate $q_\pi(s,a)$, the expected return when starting in state $s$, taking action $a$, and thereafter following policy $\pi$. Esentially, this is very similar as only for state values, except now we talk about visitis to *state-action* paris, rather than just state alone. The only problem if we do so, is that many state-action pairs may never be visited due to the sheer amount of possibilities, depending on the problem. Moreover, the $\pi$ is deterministic, then we will only observe returns for on of the actions from each state and the estimates of other actions will not improve with experience. The general goal to solving this problem is to *maintain exploration* by assuring it is happening on regular basis. Single way to do it is to specify all possibilities of state-action pairs at the start of the episode and give them non-zero probability of being selected at the start, which ensures infinite visits in the limit of an infinite number of episodes. This assumption is called *exploring starts*.

Another, more practical approach is to assure all state-action pairs are encountered by considering only policies that are stochastic with a non-zero probability of selecting all actions in each state.

### Monte Carlo Control

We can now use Monte Carlo to approximate optimal policies using the *generalized policy iteration*, where we maintain both an approximate policy and an approximate value function. 

Policy improvement is done by making the policy greedy with respect to the current value function - in this case we have an *action-value* function and no model is needed to construct the greedy policy. Greedy policy is the one that, for each $s \in S$, deterministically chooses an action with maximal action-value:

$$\pi(s) = \text{argmax}_a q(s,a)$$

For Monte Carlo policy evaluation it is natural to alternate between evaluation and improvement on an episode-by-episode basis. After each episode, the observed returns are used for policy evaluation, and then the policy is improved at all the states visited in the episode. We call this algorithm *Monte Carlo with Exploring Starts*.

<img src="./resources/monte-carlo-ES.JPG" />

```python
def mc_es(env, dealer_policy=None, player_policy=None, gamma=1., n_episodes=1000):
   state_size = (env.goal+1, max(env.available_cards)+1)
   action_size = len(env.available_actions),
   
   policy = np.zeros(state_size)
   Q = np.zeros(state_size + action_size)
   returns = defaultdict(list)
   
   for i_episode in range(n_episodes):
      # Generate an episode starting from empty state and store intermediate results in list
      state, reward, done = env.reset()
      episode_steps = []
      if done: episode_steps.append((state, 1, reward))
      else:
         initial_action = True
         while not done:
            # Initial action is random, else generate action from policy
            if initial_action: action = None
            else: action = policy[state[0], state[1]]
               
            next_state, action, reward, done = env.step(player_policy=action, dealer_policy=dealer_policy)
            episode_steps.append((state, action, reward))
            state = next_state
            initial_action = False

      # Iterate through the episode steps list in reversed order and calculate expcted returns
      G = 0
      for episode_step in reversed(episode_steps):
         state, action, reward = episode_step
         G = gamma * G + reward
         returns[state, action].append(G)
         Q[state[0], state[1], action] = np.average(returns[state, action])
         policy[state[0], state[1]] = np.argmax(Q[state[0], state[1]])
         
   return Q, policy, returns
```

```python
bj = Blackjack()

dealer_policy = lambda x: 1 if x >= 17 else 0
```

```python
Q, policy, returns = mc_es(bj, dealer_policy, n_episodes=250000)
```

```python
import matplotlib.patches as mpatches

plt.imshow(policy, cmap='Greys')
plt.xlim(1, 11)
plt.ylim(11, 21)
plt.xlabel('Dealer showing')
plt.ylabel('Player sum')

colors = ('black', 'white')
labels = ('stick', 'hit')

patches = [mpatches.Patch(color=color, label=label) for color, label in zip(colors, labels)]
plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0. )
plt.show()
```

### Monte Carlo Control without Exploring Starts

The only way to ensure that all actions are selected infinitely often is for the agent to continue to select them. There are two approaches that can help us ensure this:

- *on-policy* - methods that attempt to evaluate or improve the policy that is used to make decisions
- *off-policy* - methods the evaluate of improve a policy different from that used to generate the data.

Monte Carlo ES is an example of on-policy method. Such methods are generally *soft*, meaning that $\pi(a|s) > 0$ for all $s \in S$ and all $a \in A(s)$, but gradually shifted closer and closer to deterministic optimal policy. We can use *$\epsilon$-greedy* policies to make an on-policy MC method, meaning that most of the time they choose an action that has maximal estimated action value, but with probability $\epsilon$ they instead select an action at random. Thus, all nongreedy actions are given the minimal probability of selection $\frac{\epsilon}{|A(s)|}$, and the remaining remaining probability of given the the greedy action.

<img src="./resources/MC-eps-soft.JPG" />

```python
def mc_eps(env, dealer_policy, epsilon=0.1, gamma=1., n_episodes=1000):
   state_size = (env.goal+1, max(env.available_cards)+1)
   action_size = len(env.available_actions),
   
   # Initialize arbitrary policy where all actions have 50% chance of being selected
   policy = np.full(state_size + action_size, 0.5)
   Q = np.zeros(state_size + action_size)
   returns = defaultdict(list)
   
   for i_episode in range(n_episodes):
      # Generate an episode starting from empty state and store intermediate results in list
      state, reward, done = env.reset()
      episode_steps = []
      if done: episode_steps.append((state, 1, reward))
      else:
         while not done:
            # Generate an action from a given policy
            action = np.random.choice(env.available_actions, p=policy[state[0], state[1]])
               
            next_state, action, reward, done = env.step(player_policy=int(action), dealer_policy=dealer_policy)
            episode_steps.append((state, action, reward))
            state = next_state

      # Iterate through the episode steps list in reversed order and calculate expcted returns
      G = 0
      for episode_step in reversed(episode_steps):
         state, action, reward = episode_step
         G = gamma * G + reward
         returns[state, action].append(G)
         Q[state[0], state[1], action] = np.average(returns[state, action])
         
         greedy_action = np.argmax(Q[state[0], state[1]])
         for a in env.available_actions:
            if a == greedy_action:
               policy[state[0], state[1], a] = 1. - epsilon + epsilon/len(env.available_actions)
            else:
               policy[state[0], state[1], a] = epsilon/len(env.available_actions)
         
   return Q, policy, returns
```

```python
Q, policy, returns = mc_eps(env, dealer_policy, epsilon=0.05, n_episodes=500000)
```
